```python
import pandas 
import pyarrow.parquet as pq
import requests
import io
from pathlib import Path

root_marker = 'README.md'
current_dir = Path.cwd()
while not (current_dir / root_marker).exists() and current_dir != current_dir.parent:
    current_dir = current_dir.parent
```

```python
x = requests.get('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-01.parquet')

parquet_buffer = io.BytesIO(initial_bytes=x.content)

table = pq.read_table(source=parquet_buffer)

dataframe = pandas.DataFrame()
dataframe = table.to_pandas()   
df_sample = dataframe.iloc[:100000].copy()   
 
df_sample.to_excel(f"{current_dir}/data/raw/sample_yellow_tripdata_2024-01.xlsx", index=False)
del dataframe
```
